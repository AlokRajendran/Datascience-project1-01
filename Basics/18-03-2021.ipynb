```python
#Multiple plotting in bar chart
```

```python
x_val=['anu','manu','job','sherin','louis','manu']
x_range=np.arange(len(x_val))
y_val=[55,45,67,50,48,47]
width=1
plt.bar(x_range-width,y_val,width=width,label='current weight')
y_val2=[42,34,55,67,66,60]
plt.bar(x_range,y_val2,width=width,label='weight needed')
plt.title("Person Age Diagram")
plt.xlabel("Name")
plt.ylabel("Weight")
plt.xticks(ticks=x_range,labels=x_val)
plt.legend()
plt.grid(True)
plt.style.use("seaborn-muted")
plt.show()
```

```python
#Custom colors and Wedge Properties
```

```python
slices=[30,40,20]
labels=['men','women','kid']
colors=['red','blue','yellow']
plt.title("My Pie Chart")
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'linewidth':3,"edgecolor":"c"})
plt.show()
```

```python
#Explodes and Shadow
```

```python
slices=[30,40,20]
labels=['men','women','kid']
colors=['red','blue','yellow']
explodes=[0,0,.2]
plt.title("My Pie Chart")
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'linewidth':3,"edgecolor":"c"},explode=explodes)
```

```python
slices=[30,40,20]
labels=['men','women','kid']
colors=['red','blue','yellow']
explodes=[0,0,.2]
plt.title("My Pie Chart")
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'linewidth':3,"edgecolor":"c"},explode=explodes,shadow=True)
```

```python
#Startangle and Percenatge Value
```

```python
slices=[30,40,20]
labels=['men','women','kid']
colors=['red','blue','yellow']
explodes=[0,0,.2]
plt.title("My Pie Chart")
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'linewidth':3,"edgecolor":"c"},startangle=190,
        explode=explodes,shadow=True)
```

```python
slices=[30,40,20]
labels=['men','women','kid']
colors=['red','blue','yellow']
explodes=[0,0,.2]
plt.title("My Pie Chart")
plt.pie(slices,labels=labels,colors=colors,wedgeprops={'linewidth':3,"edgecolor":"c"},startangle=190,autopct='%1.3f%%',
        explode=explodes,shadow=True)
```
