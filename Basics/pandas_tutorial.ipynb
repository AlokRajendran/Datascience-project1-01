#sorting dataframes

```python
#Two useful functions
#sort_values():sort dataframe by one or more coulmns
#sort_index(): sort dataframe by  raw index
```

```python
#sort by year
sorted_data=df.sort_values(by="Outlet_Establishment_Year")
#print sorted data
sorted_data[:5]
```

sort values takesplace multiple options
ascending:The default sorting order is ascending,when you pass false here
inplace:whether to do inplace sorting or not

```python
#sort in place and descending order
```

```python
df.sort_values(by="Outlet_Establishment_Year",ascending=False,inplace=True)
df[:5]
```

sort a dataframe,baesd on the values of multiple columns.wwe can specify the columns we want to sort by a list in the argument for sort_values() 

```python
#read the dataset
```

```python
df1=pd.read_csv("bigmart_data.csv")
#drop the null values
df1=df1.dropna(how="any")
#sort by multiple column
df1.sort_values(by=["Outlet_Establishment_Year","Item_Outlet_Sales"],ascending="false")[:5]
```
